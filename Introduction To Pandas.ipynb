```python
load_df=pd.read_csv('https://raw.githubusercontent.com/Explore-AI/Public-Data/master/Data/fundamentals/football_players.csv', index_col='Name')
df=pd.DataFrame(load_df)
df.head()
```

#### Accessing DataFrame


To access by index only in a DataFrame we can use the <b>*iloc* or *loc* functions</b> with the indices in square brackets. The iloc function refers to the index location, so we pass in the number of the index, while the loc function refers to the name of the index, so we pass in the index name. We can use slicing if we want more than one index, for example:

dataframe.iloc[index i] – returns series at index i
dataframe.iloc[index start: index end] – returns DataFrame from start to end (end not included)
dataframe.loc['index name'] – returns series of given index name

```python
df.iloc[0:4]
```

```python
df.loc["L. Messi"
       ]
```
